```python
def cleaning(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    return text
```

```python
df['cleaned_data']=df['Data'].apply(cleaning)
df['cleaned_data']
```

```python
# remove stop word:
df['no_sw'] = df['Data'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df['no_sw']
```

```python
df['no_sw'].head()
```

```python
#Feature extraction using TF-IDF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer
```

```python

tfidf = vectorizer.fit_transform(df['no_sw'])
pd.DataFrame(tfidf.toarray())
```

```python
#splitting the data into training and testing data
from sklearn.model_selection import train_test_split
```

```python
X=tfidf
y=df['Labels']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=30)
```

```python
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import classification_report, confusion_matrix
CNB = ComplementNB()
CNB.fit(X_train, y_train)
```

```python
from sklearn import metrics
predicted = CNB.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)
```

```python
print('ComplementNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
```

```python
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
from sklearn.naive_bayes import MultinomialNB

MNB = MultinomialNB()
MNB.fit(X_train, y_train)
```

```python
predicted = MNB.predict(X_test)
accuracy_score = metrics.accuracy_score(predicted, y_test)
```

```python
print('MultinominalNB model accuracy is',str('{:04.2f}'.format(accuracy_score*100))+'%')
print('------------------------------------------------')
```

```python
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
#sentiment Analysis
sia=SentimentIntensityAnalyzer()
```

```python
sentiments=[]
for text in df['Data']:
  sentiment_scr=sia.polarity_scores(text)
  if sentiment_scr['compound']>0:
    sentiments.append('Positive+ve')
  elif sentiment_scr['compound']<0:
    sentiments.append('Negative-ve')
  else:
      sentiments.append('Neutral')
```

```python
df['sentiment']=sentiments
```

```python
final_df=df.groupby(['Labels','sentiment']).size().unstack(fill_value=0)
final_df
```
